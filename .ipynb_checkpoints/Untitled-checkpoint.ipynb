```python
HORIZON = 1 
WINDOW_SIZE = 7
```

```python
test_data = pd.concat([train_data[-WINDOW_SIZE:], test_data])
```

```python
scaler = MinMaxScaler()
scaler.fit(train_data)
scaled_train_data = scaler.transform(train_data)
scaled_test_data = scaler.transform(test_data)
```

```python
def get_labelled_windows(x, horizon=1):
    return x[:, :-horizon], x[:, -horizon:]
```

```python
def make_windows(x, window_size=7, horizon=1):
    window_step = np.expand_dims(np.arange(window_size + horizon), axis = 0)
    window_indexes = window_step + np.expand_dims(np.arange(len(x) - (window_size + horizon - 1)), axis = 0).T
    windowed_array = x[window_indexes]
    windows, labels = get_labelled_windows(windowed_array, horizon=horizon)
    return windows, labels
```

```python
train_windows, train_labels = make_windows(scaled_train_data, window_size=WINDOW_SIZE, horizon=HORIZON)
test_windows, test_labels = make_windows(scaled_test_data, window_size=WINDOW_SIZE, horizon=HORIZON)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch

def build_model(hp):
    model = keras.Sequential()
    
    # Define the number of LSTM layers
    num_layers = hp.Int('num_layers', min_value=1, max_value=3)
    
    # Add LSTM layers
    for i in range(num_layers):
        model.add(layers.LSTM(units=hp.Int(f'lstm_units_{i}', min_value=32, max_value=256, step=32),
                              return_sequences=(i != num_layers - 1),
                              input_shape=(10, 1)))
    
    model.add(layers.Dense(units=1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# Define the hyperparameter search space
tuner = RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=5,
    executions_per_trial=1,
    directory='my_dir',
    project_name='lstm_tuning'
)

# Perform hyperparameter search
tuner.search(X_train, y_train, epochs=10, validation_data=(X_val, y_val))

# Get the best hyperparameters and evaluate on the validation set
best_hyperparameters = tuner.get_best_hyperparameters()[0]
best_model = tuner.hypermodel.build(best_hyperparameters)
best_model.fit(X_train, y_train, epochs=10, validation_data=(X_val, y_val))
```
