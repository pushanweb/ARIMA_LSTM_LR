```python
HORIZON = 1 
WINDOW_SIZE = 30
```

```python
test_data = pd.concat([train_data[-WINDOW_SIZE:], test_data])
val_size = int(len(train_data) * 0.2)
validation_data = train_data[-val_size:]
train_data = train_data[:-val_size]
```

```python
scaler = MinMaxScaler()
scaler.fit(train_data)
scaled_train_data = scaler.transform(train_data)
scaled_validation_data = scaler.transform(validation_data)
scaled_test_data = scaler.transform(test_data)
```

```python
def get_labelled_windows(x, horizon=1):
    return x[:, :-horizon], x[:, -horizon:]
```

```python
def make_windows(x, window_size=7, horizon=1):
    window_step = np.expand_dims(np.arange(window_size + horizon), axis = 0)
    window_indexes = window_step + np.expand_dims(np.arange(len(x) - (window_size + horizon - 1)), axis = 0).T
    windowed_array = x[window_indexes]
    windows, labels = get_labelled_windows(windowed_array, horizon=horizon)
    return windows, labels
```

```python
train_windows, train_labels = make_windows(scaled_train_data, window_size=WINDOW_SIZE, horizon=HORIZON)
valid_windows, valid_labels = make_windows(scaled_validation_data, window_size=WINDOW_SIZE, horizon=HORIZON)
test_windows, test_labels = make_windows(scaled_test_data, window_size=WINDOW_SIZE, horizon=HORIZON)
```

```python
train_windows.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from kerastuner.tuners import RandomSearch

def build_model(hp):
    model = keras.Sequential()
    
    # Define the number of LSTM layers
    num_layers = hp.Int('num_layers', min_value=1, max_value=3)
    
    # Add LSTM layers
    for i in range(num_layers):
        model.add(layers.LSTM(units=hp.Choice(f'lstm_units_{i}', [32,64,128,256,512]),
                              return_sequences=(i != num_layers - 1),
                              input_shape=(WINDOW_SIZE,HORIZON)))
    
    model.add(layers.Dense(units=1))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model

# Define the hyperparameter search space
tuner = RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=30,
    executions_per_trial=1,
    directory='my_dir',
    project_name='original_tuning'
)

# Perform hyperparameter search
tuner.search(train_windows, train_labels, epochs=20, validation_data = (valid_windows, valid_labels))
tuner.results_summary()
```

```python
# Get the best hyperparameters and evaluate on the validation set
best_hyperparameters = tuner.get_best_hyperparameters()[0]
best_model = tuner.hypermodel.build(best_hyperparameters)
history = best_model.fit(train_windows, train_labels, epochs=100, validation_data=(valid_windows, valid_labels))
val_loss_per_epoch = history.history['val_loss']
best_epoch = val_loss_per_epoch.index(min(val_loss_per_epoch)) + 1
```

```python
best_model = tuner.hypermodel.build(best_hyperparameters)
# Retrain the model
history_best_model = best_model.fit(train_windows, train_labels, epochs=best_epoch, validation_data=(valid_windows, valid_labels))
best_model.summary()
```

```python
scaled_predictions = list()
batch = scaled_train_data[-WINDOW_SIZE:]
current_batch = batch.reshape((1,WINDOW_SIZE,1))

for i in range(293):
    pred = best_model.predict(current_batch)[0]
    scaled_predictions.append(pred)
    current_batch = np.append(current_batch[:,1:,:],[[pred]],axis=1)

predicted_prices = scaler.inverse_transform(np.array(scaled_predictions))
```
